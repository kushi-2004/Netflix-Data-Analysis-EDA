```python
df['year_added'] = pd.to_datetime(df['date_added']).dt.year
content_by_year = df.groupby(['year_added', 'type']).size().unstack().fillna(0)
plt.figure(figsize=(12,6))
sns.lineplot(data=content_by_year, markers=True)

plt.title('Content Added over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Content')
plt.grid(True)
plt.show()
```

```python
df['date_added'] = df['date_added'].str.strip()
df['year_added'] = pd.to_datetime(df['date_added'], format='mixed').dt.year

content_by_year = df.groupby(['year_added', 'type']).size().unstack().fillna(0)
plt.figure(figsize=(12,6))
sns.lineplot(data=content_by_year, markers=True, linewidth=2.5)

plt.title('The Netflix Explosion: Content Added Over Years', fontsize=15)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Titles', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Content Type')

plt.show()
```
